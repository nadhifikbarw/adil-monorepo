# Convert JSON Data

```python
# Load json data and turn into dictionaries
import json

with open('peraturan.json', 'r') as f:
    data = json.loads(f.read())
data
```

```python
# nornmalized all
normal_df = pd.json_normalize(data)
# Enlarge column width
normal_df.head().style.set_properties(subset=['url'], **{'width': '500px'})
```

### Category Columns

```python
normal_df['category']
```

```python
empty_category = normal_df['category'].apply(lambda x: True if len(x) == 0 else False)
```

```python
empty_category[empty_category]
```

```python
empty_category[empty_category]
```

```python
# Number of no category
empty_category[empty_category].sum()
```

```python
# Validate if the category is empty
normal_df.loc[26]
```

```python
# Fill the empty category
def fill_empty_cat(x):
    if len(x) == 0:
        x.append('No Category')
        return x
    else:
        return x

normal_df['category'] = normal_df['category'].apply(lambda x: fill_empty_cat(x))
```

```python
normal_df['category']
```

```python
category_list = []
for c in normal_df['category']:
    for x in c:
        category_list.append(x)
cat_series = pd.Series(category_list)
```

```python
cat_series
```

```python
cat_series.nunique()
```

```python
pd.set_option("display.max_rows", 20)
cat_series.value_counts().to_frame()
```

```python
cat_series_w_cat = cat_series[cat_series != 'No Category']
plt.figure(figsize=(12, 15))
plot = sns.countplot(y=cat_series_w_cat).set_title('Categories Count')
plt.tight_layout()
```

### Tanggal Diterapkan Column

```python
# initiate the series
tanggal = normal_df['tgl_ditetapkan']
tanggal
```

```python
# find NaN values
tanggal[tanggal.isna()]
```

```python
type(tanggal[0])
```

```python
# Fill in missing data
def fill_missing_date(x):
    if x is None:
        return 'Tanggal tidak tersedia'
    else:
        return x

normal_df['tgl_ditetapkan'] = tanggal.apply(lambda x: fill_missing_date(x))
```
